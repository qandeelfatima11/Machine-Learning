## Perform train-test split and then perform transformation

```python
df.head()
```

```python
x = df.drop('disease', axis = 1)
x
```

```python
y = df['disease']
y
```

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size =0.2, random_state=99)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

## One hot encoding on gender and city

```python
one = OneHotEncoder(drop='first', sparse_output = False, dtype = np.int32)
one.fit(X_train[['gender', 'city']])
gender_city_train = one.transform(X_train[['gender', 'city']])
gender_city_test  = one.transform(X_test[['gender', 'city']])
gender_city_train.shape, gender_city_test.shape
```

## Applying Ordinal Encoder to bp and cough column

```python
oe  = OrdinalEncoder(categories=[['low', 'normal', 'high'], ['Mild', 'Moderate', 'Strong']], dtype = np.int8)
oe.fit(X_train[['bp', 'cough']])
X_train_bp_cough_arr = oe.transform(X_train[['bp', 'cough']])
X_test_bp_cough_arr = oe.transform(X_test[['bp', 'cough']])
X_train_bp_cough_arr.shape, X_test_bp_cough_arr.shape
```

## Applying Label Encoder on disease column

```python
le = LabelEncoder()
le.fit(Y_train)
y_train = le.transform(Y_train)
y_test = le.transform(Y_test)
y_train.shape, y_test.shape
```

```python
y_train
```

## Apply simple imputer to Age column

```python
df.info()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values = np.nan, strategy = 'mean')

si.fit(X_train[['age']])
x_train_age = si.transform(X_train[['age']])
x_test_age = si.transform(X_test[['age']])
x_train_age.shape, x_test_age.shape
```

## Combine

```python
X_trained_transformed = np.concatenate((X_train_bp_cough_arr, x_train_age,  gender_city_train ), axis = 1)
X_trained_transformed.shape
```

```python
X_test_transformed = np.concatenate((X_test_bp_cough_arr, x_test_age,  gender_city_test ), axis = 1)
X_test_transformed.shape
```

## Using Column transformer to Impute

```python
ohe = OneHotEncoder(drop = 'first', sparse= False, dtype = np.int32)

oe = OrdinalEncoder(categories = [['low', 'normal', 'high'], ['Mild', 'Moderate', 'Strong']], dtype = np.int8)

si = SimpleImputer(missing_values = np.nan, strategy='mean')
```

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers= [
    ('tfm1' , ohe, ['gender', 'city']),
    ('tfm2', oe , ['bp', 'cough']),
    ('tfm3', si, ['age'])
    ], remainder='passthrough')
```

```python
transformer.fit(X_train)

X_trained_transformed1 = transformer.transform(X_train)
X_test_transformed1 = transformer.transform(X_test)

X_trained_transformed_df1 = pd.DataFrame(X_trained_transformed1, columns=['gender', 'city1', 'city2', 'city3', 'age', 'bp', 'cough'])
X_trained_transformed_df1
```
